## 1. Variables and Data Types

```python
# Integers
shares_owned = 100

# Floats 
stock_price = 150.75

# Strings
ticker = "AAPL"

# Booleans
is_profitable = True

print(f"Ticker: {ticker}, Price: ${stock_price}, Shares: {shares_owned}")
```

## 2. Arithmetic Operators

```python
buy_price = 100.00
sell_price = 120.00
shares = 50

profit = (sell_price - buy_price) * shares
return_pct = ((sell_price - buy_price) / buy_price) * 100

print(f"Profit: ${profit}")
print(f"Return: {return_pct}%")

# Compound interest
principal = 10000
rate = 0.07
years = 10
future_value = principal * (1 + rate) ** years
print(f"\n${principal} at {rate*100}% for {years}yr = ${future_value:.2f}")
```

## 3. Comparison & Logical Operators

```python
price = 155.00
target = 180.00
stop_loss = 140.00

print(f"Above stop loss: {price > stop_loss}")
print(f"Below target: {price < target}")

# Logical operators
rsi = 35
macd_bullish = True

buy_signal = (rsi < 40) and macd_bullish
print(f"\nBuy signal: {buy_signal}")
```

## 4. String Operations

```python
ticker = "MSFT"
company = "Microsoft Corporation"

print(f"Full: {ticker} - {company}")
print(f"Upper: {company.upper()}")
print(f"Length: {len(company)}")

# f-string formatting
price = 378.91
change = 2.45
print(f"\n{ticker}: ${price:.2f} (+${change:.2f})")
```
